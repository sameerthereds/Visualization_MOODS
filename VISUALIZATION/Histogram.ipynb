```python
df=pd.read_csv("Stressors_day .csv")
df=df.sort_values(by=['Day'])
print(df.head(5))
x_var = 'Day'
groupby_var = 'Stressors'
df_agg = df.loc[:, [x_var, groupby_var]].groupby(groupby_var)
vals = [df[x_var].values.tolist() for i, df in df_agg]

# Draw
fig, ax = plt.subplots(figsize=(16,9), dpi= 100)
#plt.figure(figsize=(16,9), dpi= 100)
colors = [plt.cm.Spectral(i/float(len(vals)-1)) for i in range(len(vals))]
n, bins, patches = ax.hist(vals, df[x_var].unique().__len__(), stacked=False,rwidth=0.4, density=False, color=colors[:len(vals)])

# Decoration
plt.legend({group:col for group, col in zip(np.unique(df[groupby_var]).tolist(), colors[:len(vals)])})
plt.title(f" Histogram of ${x_var}$ colored by ${groupby_var}$", fontsize=22)
# plt.xlabel(x_var)
plt.ylabel("Frequency")
plt.ylim(0, 5)
labels=['Monday', 'Tuesday', 'Wednesday','Thursday', 'Friday', 'Saturday','Sunday']
#plt.xticks(ticks=bins, labels=['Monday', 'Tuesday', 'Wednesday','Thursday', 'Friday', 'Saturday','Sunday'], rotation=90, horizontalalignment='right')
ax.set_xticklabels(labels, rotation=45, ha='right')
# plt.setp( ax.xaxis.get_majorticklabels(), rotation=-45, ha="right" )
plt.savefig("hist_stressors.png")
plt.show()
```
