```python
import plotly.express as px
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("week of stream_graph - Sheet1.csv")
```

```python
df=data.pivot_table(index='Week', columns='Stressors', values='Count')
df=df.fillna(0)
df=df.reset_index()
#df["Week"]=["1","2","3","4","5","6","7"]
```

```python
df
```

```python
fig = px.parallel_coordinates(df, color="Stressors", labels={"Stressors": "Stressors",
                1: "Week 1", 2: "Week 2",
                3: "Week 3", 4: "Week 4",  5: "Week 5", 6: "Week 6",
                7: "Week 7",},
                             color_continuous_scale=px.colors.diverging.Tealrose,
                            )
fig.show()
```

```python
fig  = plt.figure(figsize=(10,10),facecolor = 'yellow', dpi=100) 
pd.plotting.parallel_coordinates(
    df, 'Week', color=('#556270', '#4ECDC4', '#C7F464',"#802200","#B33000","#FF4500","#FAA327")
)
plt.savefig("parralel.png")
```
