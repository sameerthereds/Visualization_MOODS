```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
data=pd.read_csv("stream_graph - Sheet1.csv")
```

```python
cols=data.Stressors.unique()
x=range(33)
dict_count={k: list(v) for k,v in data.groupby("Stressors")["Count"]}
temp=[]
for i in dict_count:
    temp.append(dict_count[i])
temp_mod=[]
for i in temp:
    temp_len=len(i)
    to_add=33-temp_len
    if to_add > 0:
        i=i + [0] * to_add
    temp_mod.append(i)
```

```python
fig  = plt.figure(figsize=(10,10),facecolor = 'yellow', dpi=100)
plt.stackplot(x,temp_mod,labels=["Transportation","Work/School","Finance","Friends/Family","Health","Personal",
                            "Social Media"])
plt.legend(loc='upper right')

plt.savefig("stacked_area.png")
```
